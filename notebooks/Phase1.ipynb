### Task 1.2: Use the datasets library from Hugging Face to download the arabic-generated-abstracts dataset directly into a Python environment (e.g., Jupyter Notebook, Google Colab).

```python
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
splits = {'by_polishing': 'data/by_polishing-00000-of-00001.parquet', 'from_title': 'data/from_title-00000-of-00001.parquet', 'from_title_and_content': 'data/from_title_and_content-00000-of-00001.parquet'}
polishing = pd.read_parquet("hf://datasets/KFUPM-JRCAI/arabic-generated-abstracts/" + splits["by_polishing"])
title = pd.read_parquet("hf://datasets/KFUPM-JRCAI/arabic-generated-abstracts/" + splits["from_title"])
title_content = pd.read_parquet("hf://datasets/KFUPM-JRCAI/arabic-generated-abstracts/" + splits["from_title_and_content"])
```

#### second conatenate the second and thirsd files & concatenate the llm responses in them

```python
title['Generated_abstract'] = title.iloc[:, -4:].apply(lambda x: ''.join(x.dropna().astype(str)), axis=1)
title_content['Generated_abstract'] = title_content.iloc[:, -4:].apply(lambda x: ''.join(x.dropna().astype(str)), axis=1)
```

```python
title= title.drop(columns = ['allam_generated_abstract', 'jais_generated_abstract', 'llama_generated_abstract', 'openai_generated_abstract'])
title_content = title_content.drop(columns = ['allam_generated_abstract', 'jais_generated_abstract', 'llama_generated_abstract', 'openai_generated_abstract'])
```

```python
merged = pd.merge(title, title_content, on='original_abstract', how='outer')
```

```python
merged.columns = ['original_abstract', 'generated_title', 'generated_title_content']
```

```python
merged['Generated_abstract'] = merged.iloc[:, -2:].apply(lambda x: ''.join(x.dropna().astype(str)), axis=1)
```
